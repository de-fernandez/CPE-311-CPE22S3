```python
from statistics import mean, median, mode, variance, stdev
```

Mean

```python
# To find the mean with statistics module
mean = mean(salaries)
mean
```

```python
# To find the mean without statistics module manually by dividing the summation of salaries by the number of salaries in a list
mean_data = sum(salaries)/len(salaries)
mean_data
```

Median

```python
# Sort first the list in order to find the correct middle number of the list
salaries = sorted(salaries)
```

```python
# To find the median with statistics module
median(salaries)
```

```python
# To find the median without statistics module manually by counting first the number of the list and dividing it by half,
# and have the result of it as the parameter for the list
median_number = len(salaries)
median_number = median_number/2
median_data = salaries[int(median_number)]
median_data
```

Mode

```python
# To find the mode with statistics module
mode(salaries)
```

```python
# To find the mode without statistics module, by having the collections module
from collections import Counter

counter = Counter(salaries) # This count the frequency of each number in the list
mode_data = counter.most_common(1)[0][0] # The ".most_common" picks out the highest frequency among the salaries in the list
mode_data
```

Sample Variance

```python
# To find the sample variance with statistics module
variance(salaries, mean)
```

```python
# To find the sample variance without statistics module, by having pandas
summ = sum(salaries) # This sums up the salaries
numm = len(salaries) # This counts up the number of salaries
mean_number = summ / numm # This compute for the mean of the salaries with the summ and numm that was declared earlier
variance_data = sum((x - mean_number) ** 2 for x in salaries) / (numm - 1) # The actual formula of sample variance in python
variance_data
```

Sample Standard Deviation

```python
# To find the sample standard deviation with statistics module
stdev(salaries)
```

```python
# To find the sample standard deviation without statistics module manually with the actual formula
summ = sum(salaries) # This sums up the salaries
numm = len(salaries) # This counts up the number of salaries
mean_number = summ / numm # This compute for the mean of the salaries with the summ and numm that was declared earlier
deviation_number = sum((x - mean_number) ** 2 for x in salaries) / (numm - 1) # The actual formula of sample variance in python
deviation_data = deviation_number ** (1/2) # If you square rooted the sample variance, you get the sample standard deviation. In mathematical sense, the "√x" is equal to "x^(1/2)"
deviation_data
```

 **Exercise 2**
 
Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:
- Range
- Coefficient of variation Interquartile range
- Quartile coefficient of dispersion

```python
# Write a comment per statistical function
```

```python
from statistics import mean, stdev, quantiles
```

Range

```python
# To find the range with statistics module
# The range is found by the maximum value subtracted by the minimum value

range_data = max(salaries) - min(salaries)
range_data
```

Coefficient of variation Interquartile range

```python
# To find the coefficient of variation Interquartile range with statistics module
# The coefficient variation is found by the standard deviation divided by the mean, and the result is multiplied by 100
# The interquartile range is found by the third quartile subtracted by the first quartile

co_variation = (stdev(salaries) / mean(salaries)) * 100
q1, q2, q3 = quantiles(salaries, n = 4)
q_range = q3 - q1
print(f"The Coefficient Variation is {co_variation}")
print(f"The Interquartile Range is {q_range}")
```

Quartile coefficient of dispersion

```python
# To find the quartile coefficient of dispersion with statistics module
# The quartile coefficient of dispersion is found by the third quartile subtracted by the first quartile,
# divided by the result of the summation of the third quartile and the first quartile

q1, q2, q3 = quantiles(salaries, n = 4)
qc_dispersion = (q3 - q1) / (q3 + q1)
qc_dispersion
```
