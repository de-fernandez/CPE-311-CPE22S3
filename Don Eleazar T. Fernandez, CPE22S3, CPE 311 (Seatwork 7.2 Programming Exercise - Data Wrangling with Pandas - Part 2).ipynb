```python
data = pd.read_csv("Meteorite_Landings.csv")
meteorites = pd.DataFrame(data)
meteorites.head()
```

1. Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to convert the year column to a number as we did in the previous exercise.


```python
filter = lambda x: x.str[6:10]
meteorites["year"] = filter(meteorites["year"])
meteorites["year"] = meteorites["year"].apply(pd.to_numeric)
meteorites.dtypes
```

```python
meteorites_filtered = meteorites[(meteorites["year"] >= 2005) & (meteorites["year"] <= 2009)]
```

```python
meteorites_filtered = meteorites_filtered.set_index("year")
```

```python
meteorites_Fell = meteorites_filtered[(meteorites_filtered["fall"] == "Fell")]
meteorites_Fell.iloc[:, 4].quantile([0.95])
```

```python
meteorites_Found = meteorites_filtered[(meteorites_filtered["fall"] == "Found")]
meteorites_Found.iloc[:, 4].quantile([0.95])
```

```python
Number_1_1 = meteorites_filtered.groupby(["year", "fall"])[["mass (g)"]].quantile(0.95)
Number_1_1
```

```python
Number_1_2 = meteorites_filtered.groupby(["year", "fall"])[["name"]].count()
Number_1_2
```

```python
Number_1_1["95th Percentile"] = Number_1_1
Number_1_1["Count"] = Number_1_2
Number_1_1
```

```python
data1 = pd.read_csv("2019_Yellow_Taxi_Trip_Data.csv")
taxi = pd.DataFrame(data1)
taxi.head()
```

1. Using the taxi trip data in the 2019_Yellow_Taxi_Trip_Data.csv file, resample the data to an hourly frequency based on the dropoff time. Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.

```python
# resample the data to an hourly frequency based on the dropoff time
taxi['tpep_dropoff_datetime'] = pd.to_datetime(taxi['tpep_dropoff_datetime'])
taxi.set_index("tpep_dropoff_datetime", inplace = True)
taxi.index = pd.to_datetime(taxi.index)
taxi['Hour'] = taxi.index.hour

# Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.
taxi = taxi.groupby('Hour')[['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']].sum()
taxi = taxi["tip_amount"].nlargest(5)
taxi
```
