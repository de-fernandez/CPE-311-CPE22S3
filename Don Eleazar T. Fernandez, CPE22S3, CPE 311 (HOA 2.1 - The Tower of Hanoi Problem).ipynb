```python
s = [] # The source list
a = [] # The auxiliary list
d = [] # The destination list
n = 3 # The number of disk that can be made

for i in range(n, 0, -1): # With the number of n (which is the disk), this create the list of disk for the number of n
  s.append(i)

class Variable(): # To initialize each of the variable
  def __init__(self, n, s, a, d):
    self.n = n
    self.s = s
    self.a = a
    self.d = d

class TowerOfHanoi(Variable):
  def Transfer(self, listA, listB): # This either add or remove a value of the list respectively according to the Tower Function
    value = listA.pop()
    listB.append(value)

  def Tower(self, n, source, auxiliary, destination): # The recursive function iterates itself by self calling for a definite number of period (by number of disk)
    if n == 1:
      self.Transfer(source, destination)
      print(f"Source: {source}, Auxiliary:{auxiliary}, Destination: {destination}") # To print
      return

    self.Tower(n - 1, source, destination, auxiliary)
    self.Transfer(source, destination)

    print(f"Source: {source}, Auxiliary:{auxiliary}, Destination: {destination}") # To print
    self.Tower(n - 1, auxiliary, source, destination)

print(f"Source: {s}, Auxiliary:{a}, Destination: {d}") # To print
tower = TowerOfHanoi(n, s, a, d)
tower.Tower(n, s, a, d) # To call out the function Tower on the class TowerOfHanoi
```
